```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import Dataset/synth_eye_medium_train_scaling.csv
df = pd.read_csv('Dataset/synth_eye_medium_train_scaling.csv')
df.head()
```

```python
# plot distribution of column M00, M01, M02, M10, M11, M12 in subplots
fig, axs = plt.subplots(2, 3, figsize=(15, 6))
axs[0, 0].hist(df['M00'], bins=50)
axs[0, 0].set_title('M00')
axs[0, 1].hist(df['M01'], bins=50)
axs[0, 1].set_title('M01')
axs[0, 2].hist(df['M02'], bins=50)
axs[0, 2].set_title('M02')
axs[1, 0].hist(df['M10'], bins=50)
axs[1, 0].set_title('M10')
axs[1, 1].hist(df['M11'], bins=50)
axs[1, 1].set_title('M11')
axs[1, 2].hist(df['M12'], bins=50)
axs[1, 2].set_title('M12')
fig.suptitle('Distribution of M00, M01, M02, M10, M11, M12')
fig.tight_layout()

plt.show()
```
