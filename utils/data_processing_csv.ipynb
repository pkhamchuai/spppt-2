```python
import os
import re
```

```python
def flatten(l):
    ''' Flatten a list of lists into one list'''
    return [item for sublist in l for item in sublist]

def filename_splitter(dir, txt):
    ''' Split the directory path and filename into a list of strings '''
    out1 = dir.split('/')
    out2 = txt.split('.')
    out2 = re.split('_', out2[0])
    # don't add the file extension
    # don't merge 2 lists because we need to keep the directory path and filename separate
    # out1.append(re.split('_', out2[0]))
    return out1, out2
```

```python
# create a function that input directory path
# fetch the image names in that directory
# add some text e.g. before, after, etc. after the third underscore in the image name
# save the image with the new name (rename the image)
def rename_images(dir_path, text):
    ''' Rename images in the directory path with the text '''
    # fetch the image names in that directory
    for dirname, _, filenames in os.walk(dir_path):
        for filename in filenames:
            if filename[-3:] == 'jpg': #  and filename[0:5] != 'output'
                # print(dirname, filename)
                dir_splitted, filename_splitted = filename_splitter(dirname, filename)
                # print(dir_splitted, filename_splitted)
                # add some text e.g. before, after, etc. after the third underscore in the image name
                filename_splitted.insert(3, text)
                # print(filename_splitted)
                # save the image with the new name (rename the image)
                new_filename = '_'.join(filename_splitted) + '.jpg'
                # print(new_filename)
                os.rename(os.path.join(dirname, filename), os.path.join(dirname, new_filename))
                # print(os.path.join(dirname, filename))
                # print(os.path.join(dirname, new_filename))
                
                
```

```python
# create a list to store the image name and its corresponding directory path

dir_file_dict = {}
dir_list = []
for dirname, _, filenames in os.walk('Dataset/Dataset-processed'):
    for filename in filenames:
        if filename[-3:] == 'jpg': #  and filename[0:5] != 'output'
            # print(dirname, filename)
            # dir_splitted, filename_splitted = filename_splitter(dirname, filename)
            
            # add the directory path, filename to the list  (their splitted versions might not be necessary)
            if dirname not in dir_file_dict:
                dir_list.append(dirname)
                dir_file_dict[dirname] = []
            dir_file_dict[dirname].append([dirname, filename])
        
        # print(os.path.join(dirname, filename))
print(f'Number of folders containing images: {len(dir_file_dict)}')
```

```python
# read the file dataset_reg_pair.csv in Dataset-processed directory
# Specify the file path
file_path = 'Dataset/dataset_reg_pair.csv'

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path)

# Print the contents of the DataFrame
print(df)
```

```python
# count total number of images
img_count = 0
# count total number of pairs
pair_count = 0

# show how many images are in each directory
# show how many pairs can be made from the images in each directory
# pairing all images in each directory together
for key in dir_file_dict:
    # print(f'{key} has {len(dir_file_dict[key])} images = {int(len(dir_file_dict[key])*(len(dir_file_dict[key])-1)/2)} pairs')
    img_count += len(dir_file_dict[key])
    pair_count += int(len(dir_file_dict[key])*(len(dir_file_dict[key])-1)/2)    
    
print(f'Total number of images: {img_count}')
print(f'Total number of pairs: {pair_count}')
```

```python
# find image names that is duplicated in the dataset

# file name list
file_name_list = []

# find duplicated image names
for key in dir_file_dict:
    for i in range(len(dir_file_dict[key])):
        file_name_list.append(dir_file_dict[key][i][1])

# print(file_name_list)

# find duplicated image names
duplicated_file_name_list = []
for i in range(len(file_name_list)):
    if file_name_list[i] in file_name_list[i+1:]:
        duplicated_file_name_list.append(file_name_list[i])

print(f'Total number of duplicated images: {len(duplicated_file_name_list)}')
print(f'Total number of unique images: {len(file_name_list)-len(duplicated_file_name_list)}')

```

```python
# 38 dir for test, 152 dir for train
print(int(len(dir_file_dict)*0.2), int(len(dir_file_dict)*0.8))
print(int(len(dir_file_dict)*0.2) + int(len(dir_file_dict)*0.8))
```

```python
# write the directory path and filename of each image pair to csv file 'dataset_reg_pair.csv' (variable df)
# the csv file will be used to create the dataset for registration
# the dataset will be used to train the registration model
# the registration model will be used to register the images in the dataset

# write the directory path in column 'directory'
# first image in the pair is the source image (the image that will be transformed)
# second image in the pair is the target image (the image that the source image will be transformed to)
# the source image and the target image are in the same directory
# put the location of the source image in column 'Source image'
# put the location of the target image in column 'Target image'


# create a list to store the directory path and filename of each image pair
dir_file_pair_list = []

# assign 20% of the directory paths to the test set
# assign 80% of the directory paths to the training set
# create a list with length equal to the number of directory paths
# 20% of list elements will be 1, the remaining 80% will be 0
# randomly shuffle the list
# the first 20% of the list will be 1, the remaining 80% will be 0
# the first 20% of the directory paths will be assigned to the test set, the remaining 80% will be assigned to the training set
import random
test_train_list = [0]*38 + [1]*152
random.shuffle(test_train_list)
# number of directory
print(f'# directory: {len(test_train_list)}')
# number of 1 in test list
print(f'# test image directory: {sum(test_train_list)}')

i = 0
# iterate through the dictionary
for key in dir_file_dict:
    # get the names of the images in the directory
    img_list = dir_file_dict[key]
    # print(img_list)
    print(len(img_list))

    # pairing all images in the directory together
    for j in range(len(img_list)):
        for k in range(j+1, len(img_list)):
            print(f'pair: {img_list[j][1]}, {img_list[k][1]}')
            dir_file_pair_list.append([img_list[j][0], img_list[j][1], img_list[k][1], test_train_list[i]])

    i += 1

print(f'Finish assigning values to {i} directories')
print(f'Total number of pairs: {len(dir_file_pair_list)}')

print(dir_file_pair_list[:10])

# add data in the list into the opened pandas dataframe df in column 'Directory', 'Source image', 'Target image', 'type', 'before-after mixed'
df['Directory'] = [i[0] for i in dir_file_pair_list]
df['source'] = [i[0] + '/' + i[1] for i in dir_file_pair_list]
df['target'] = [i[0] + '/' + i[2] for i in dir_file_pair_list]
df['training'] = [i[3] for i in dir_file_pair_list]

df.head(100)
```
