```python
# generate a mockup of the data for the project
import random
import numpy as np
import matplotlib.pyplot as plt
from utils.utils1 import signaturebar_gray
```

```python
'''
1. linsplace 0-20, 100
2. random values, decreasing
3. random values, decreasing

split into 4 parts
'''

data = []
epochs = 20
for i in range(4):
    data.append(np.array([np.linspace((i)*epochs/4, (i+1)*epochs/4, 100), np.random.rand(100), np.random.rand(100)]).T)
```

```python
# create 2x2 subplots, plot the data as 4 different plots
fig, axes = plt.subplots(2, 2, figsize=(14, 8))
for i in range(4):
    plt.subplot(2, 2, i + 1)
    start = i * 25
    end = (i + 1) * 25
    plt.plot(data[i][:, 0], data[i][:, 1], label="Train Loss")
    plt.plot(data[i][:, 0], data[i][:, 2], label="Validation Loss")
    # plt.title(f"Stage {i + 1}")

    if i == 0 or i == 2:
        plt.ylabel("Loss")

    plt.xlabel(f"(Stage {i + 1}) Epochs")
    # else:
    #     plt.xticks([])

    if i == 1:
        plt.legend()
plt.suptitle("Model A, 10 epochs, etc")
plt.tight_layout()
# _ = signaturebar_gray(fig, "Model A, 10 epochs, etc")
```
