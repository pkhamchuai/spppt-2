```python
# generate a mockup of the data for the project
import random
import numpy as np
import matplotlib.pyplot as plt
from utils.utils1 import signaturebar_gray
```

```python
'''
1. linsplace 0-20, 100
2. random values, decreasing
3. random values, decreasing

split into 4 parts
'''

data = np.array([np.linspace(0, 20, 100), np.random.rand(100), np.random.rand(100)]).T
data[:10]
```

```python
# create 2x2 subplots, plot the data as 4 different plots
fig, axes = plt.subplots(2, 2, figsize=(10, 6))
for i in range(4):
    plt.subplot(2, 2, i + 1)
    start = i * 25
    end = (i + 1) * 25
    plt.plot(data[:, 0][start:end], data[:, 1][start:end], label="Training Loss")
    plt.plot(data[:, 0][start:end], data[:, 2][start:end], label="Validation Loss")
    # plt.title(f"Stage {i + 1}")

    if i == 0 or i == 2:
        plt.ylabel("Loss")

        plt.xlabel(f"Stage {i + 1}, Epochs")
    # else:
    #     plt.xticks([])

    if i == 1:
        plt.legend()
plt.tight_layout()
_ = signaturebar_gray(fig, "Model A, 10 epochs, etc")
```
